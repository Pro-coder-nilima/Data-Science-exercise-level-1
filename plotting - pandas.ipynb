```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [0,5,3]
y = [2,15,4]
plt.plot(x,y)
plt.title("py plot", size = '20')
plt.xlabel('y axis', size = '15')
plt.ylabel('x axis', size = '15')
plt.show()
```

```python
xpoint = np.array([0,5])
ypoint = np.array([0,8])
plt.plot(xpoint, ypoint)
plt.show()
```

### plotting without line

```python
xpoint = np.array([0,5])
ypoint = np.array([0,8])
plt.plot(xpoint, ypoint, '*')
plt.show()
```

    If we dont want line then we can give some identies for where we plotting

```python
xpoint = np.array([0,1,2,3])
ypoint = np.array([2,5,6,7])
plt.plot(xpoint, ypoint, '*')
plt.show()
```

### markers points

     Here we have various types of plotting to  use markers points.

```python
ypoints = np.array([1,2,3,4])
plt.plot(ypoints, 'o',marker = '*')
plt.show()
```

### Subplot

    Subplot means a group of smaller axes subplot divided the current figure into panes that are numbered row-wise.

```python
x = np.array([0,1,2,3])
y = np.array([0,20,30,40])

plt.subplot(1,2,1)
plt.plot (x,y,'o')

x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y,'o')

plt.show()
```

### Bar plot

    A bar plot is a plot that presents categorical data with rectangular bar with length proportional to the values that 
    they represent.

```python
languages = ["c", "c++", "Java", "python"]
student = [1,2,4,3]
plt.bar(languages, student, color = "g")
plt.show()
```

### Histogram

      A histogram is a graphical representation of data. It used to summarize discreat or continuous data that are measured
      on an interval scale.

```python
fig,ax = plt.subplots(1,1)
a = np.array([20,80,5,40,50,70,55,10,21,51,5,75,30,27])
ax.hist(a, bins = [0,25,50,74,100])
ax.set_title("Histogram", size = 25)
ax.set_xlabel("marks", size = 20)
ax.set_ylabel("No.of students", size = 20)
plt.show()
```

### pie plot

        A pie is a propotional representation of the numerical data in a column

```python
nums = [123,234,456,345,333]
labels = ['India','pakistan','china',' north','east']
explode = [0,0,0.2,0,0]
colors  = ['black','pink','blue','g','y']
plt.pie(nums,labels = labels, explode = explode, shadow=True, startangle = 90, colors = colors)
plt.title('population')
plt.grid(True)
plt.show()
```

###  Scatterplot

```python
x = [0,1,2,3]
y = [3,4,5,6]
x1 = [6,7,8,9]
y1 = [2,3,4,4]
plt.scatter(x,y,label = "High income", color = 'r')
plt.scatter(x1,y1,label = "Low income", color= "b")
plt.ylabel('saving')
plt.xlabel('income')
plt.title('scatter')
plt.legend()
plt.show()
```
