## univariate Analysis
the term "univariate" refers to single variable

## Histogram
Histogram are the most common means of looking at a single variable

```python
fig = plt.figure(figsize = (8,5))
axes1 = fig.add_subplot(1,1,1)
axes1.hist(df['total_bill'], bins = 10, color = "blue", orientation = "vertical")
axes1.set_title("Histogram of Total bills",size = 45)
axes1.set_xlabel("Frequency",size = 25)
axes1.set_ylabel("Total Bill", size = 25)
```
