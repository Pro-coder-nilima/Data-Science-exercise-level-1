```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('car_crashes')
df
```

```python
sns.distplot   #Tab+shift
```

### Distribution Plot

```python
sns.distplot(df['not_distracted'])
```

```python
sns.distplot(df['not_distracted'], kde = False, bins = 25)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (5,5))
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "kde", shade = True)
plt.suptitle("Seaborn visualition", size = 25)
```

```python
plt.figure(figsize = (5,5))
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "hex")


plt.suptitle("Seaborn visualition", size = 20)
```

### KDE plot

```python
sns.kdeplot(df['alcohol'])
```

### Pair Plots

```python
sns.pairplot(df)
```

```python
tips = sns.load_dataset('tips')
tips

sns.pairplot(tips, hue = 'sex', palette = 'Blues')
```

### Rug plot
         A rug plot is a simple plot to explore distributions of a continous variable.

```python
sns.rugplot(tips['tip'])
```

```python
sns.rugplot(data = tips, x = "total_bill")
```

```python
plt.figure(figsize = (5,5))
sns.set_style('white')
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "reg")
```

```python
plt.figure(figsize = (5,5))
sns.set_style('ticks')
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "reg")
sns.set_context('poster', font_scale = 1.5)
```

```python
tips['day']
```

### Box  Plot

```python
x = sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue= 'sex')
x
plt.legend(loc=1)
```

### Violin Plot

```python
sns.violinplot(x = 'day', y = 'total_bill', data =tips, hue ='sex')
plt.legend(loc=1)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data =tips, hue ='sex', split= True)
plt.legend(loc=1)
```

### Strip Plot

```python
plt.figure(figsize=(8,5))
sns.stripplot(x = 'day', y ='total_bill', data = tips, hue = 'sex')
```

### Swarm Plot

```python
sns.violinplot(x = 'day', y = 'total_bill',data= tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data= tips)
sns.swarmplot(x = 'day', y = 'total_bill',data= tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data= tips)
sns.swarmplot(x = 'day', y = 'total_bill',data= tips, color = 'white')
```

```python
sns.swarmplot(x ='day',y = 'total_bill',data = tips)
```

### Palettes

```python
plt.figure(figsize = (8,6))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = 'afmhot')
```

```python
plt.figure(figsize =(8,5))
sns.set_context('paper',font_scale= 1.4)

tips_mx =tips.corr()

sns.heatmap(tips_mx, annot = True, cmap = 'Blues')
```
