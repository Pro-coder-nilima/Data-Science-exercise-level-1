```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('car_crashes')
df
```

```python
sns.distplot   #Tab+shift
```

### Distribution Plot

```python
sns.distplot(df['not_distracted'])
```

```python
sns.distplot(df['not_distracted'], kde = False, bins = 25)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (5,5))
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "kde", shade = True)
plt.suptitle("Seaborn visualition", size = 25)
```

```python
plt.figure(figsize = (5,5))
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "hex")


plt.suptitle("Seaborn visualition", size = 20)
```

### KDE plot

```python
sns.kdeplot(df['alcohol'])
```

### Pair Plots

```python
sns.pairplot(df)
```

```python
tips = sns.load_dataset('tips')
tips

sns.pairplot(tips, hue = 'sex', palette = 'Blues')
```

### Rug plot
         A rug plot is a simple plot to explore distributions of a continous variable.

```python
sns.rugplot(tips['tip'])
```

```python
sns.rugplot(data = tips, x = "total_bill")
```

```python
plt.figure(figsize = (5,5))
sns.set_style('white')
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "reg")
```

```python
plt.figure(figsize = (5,5))
sns.set_style('ticks')
sns.jointplot(x = "speeding", y = "alcohol", data = df, kind = "reg")
sns.set_context('poster', font_scale = 1.5)
```
