```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = sns.load_dataset('flights')
x
```

```python

z = sns.load_dataset('flights')
sns.relplot(x = 'passengers',y = 'month', data = z)

```

```python
z = sns.load_dataset('flights')
sns.relplot(x = 'passengers',y = 'month', hue = 'year', data = z)
```

```python
b = sns.load_dataset('tips')
sns.relplot(x = 'time', y = 'tip', data = b, kind = 'line')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = b)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'boxen',  data = b)
```
